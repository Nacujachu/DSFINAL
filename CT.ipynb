```python
#抓中時、商報、旺報
from bs4 import BeautifulSoup
import requests
import time
import re
import operator
import sys
import datetime
import os

def CT(dirName,keyword,startDate,endDate):
    base = 'http://www.chinatimes.com'
    date = '/history-by-date/'
    banlist=[
        '焦點要聞',
        '政治要聞',
        '生活新聞',
        '社會新聞',
        '兩岸要聞',
        '時論廣場',
        '藝文副刊',
        '財經焦點',
        '地方新聞',
        '運動天地',
        '娛樂新聞',
        '時尚消費',
        '圖片',
        '財經要聞',
        '全球財經',
        '金融．稅務',
        '產業．科技',
        '證券．權證',
        '投資理財',
        '產業特刊',
        '焦點新聞',
        '話題觀察',
        '兩岸新聞',
        '兩岸藝文',
        '產業財經',
        '論壇廣場',
        '政治',
        '生活',
        '社會',
        '旅遊',
        '娛樂',
        '體育',
        '財經',
        '國際',
        '兩岸',
        '科技',
        '軍事',
        '健康'
    ]
    startDay = datetime.datetime.strptime(startDate,'%Y-%m-%d')
    endDay = datetime.datetime.strptime(endDate,'%Y-%m-%d')
    #print(startDay)
    #print(endDay)
    for d in range(0, (endDay-startDay).days+1):
        #print(d)
        Day = startDay.date()+ datetime.timedelta(days = d)
        Day = Day.strftime("%Y-%m-%d")
        fileName = os.path.join(dirName+'/'+keyword+Day+'.txt')
        #print(fileName)
        date = date + startDate + '-'
        all_articles = open(fileName, 'w',encoding='utf-8')
        #抓報社名稱
        for newsPaperType in range(2601,2605):
            #抓該報紙的第一頁
            res = requests.get(base+date+str(newsPaperType))
            soup = BeautifulSoup(res.text, "lxml")
            #news
            news = soup.find('div',  {'class': 'listRight'})
            try:
                news = news.find_all('a')
            except:
                continue
            #不斷抓取下一頁資訊 直到沒有下一頁
            while True:
                for i in news:
                    title = i.text.strip()
                    if title in banlist:
                        continue
                    newsURL = i.get('href')
                    #print(title)
                    #print(newsURL)
                    #all_articles.write(title + "," + base + newsURL + ",")
                    #抓新聞內容
                    newsContent = ""
                    newsRes = requests.get(base+newsURL)
                    newsSoup = BeautifulSoup(newsRes.text, "lxml")
                    text = newsSoup.find('article', {'class': 'arttext marbotm clear-fix'})
                    #print(title)
                    try:
                        text = text.find_all('p')
                    except:
                        continue
                    for i in text:
                        #print(i.text)
                        newsContent = newsContent + i.text
                    if keyword in newsContent:
                        print(base+newsURL + '---------------------------')
                        all_articles.write(title + ",")
                        all_articles.write(newsContent)
                        all_articles.write("\n")
                #抓下一頁的網址
                nextPage = soup.find('div',  {'class': 'pagination clear-fix'})
                nextPage = nextPage.find_all('a')
                nextPageURL = ""
                for i in nextPage:
                    if i.text == '下一頁':
                        nextPageURL = i.get('href')
                        #print(nextPageURL)
                        break
                #print(base+nextPageURL)
                if nextPageURL == "":
                    break
                time.sleep(0.01)
                res = requests.get(base+nextPageURL)
                #print("here!!!!!!!!!!" + base+nextPageURL)
                soup = BeautifulSoup(res.text, "lxml")
                #news
                news = soup.find('div',  {'class': 'listRight'})
                news = news.find_all('a')

        #print('-------------------------------')
        #print('屌虐')
        all_articles.close()
    
#date format : 2011-06-13
dirName = sys.argv[1]
keyword = sys.argv[2]
startDate = sys.argv[3]
endDate = sys.argv[4]
CT(dirName,keyword,startDate,endDate)
```
