```python
#抓中時、商報、旺報
from bs4 import BeautifulSoup
import requests
import time
import re
import operator
import sys

#news url base
base = 'http://www.chinatimes.com'
date = '/history-by-date/2011-06-13-'
keyword = "新加坡"
banlist=[
    '焦點要聞',
    '社會新聞',
    '兩岸要聞',
    '圖片',
    '時論廣場',
    '運動天地',
    '地方新聞',
    '藝文副刊',
    '財經要聞',
    '焦點新聞',
    '話題觀察',
    '兩岸藝文',
    '產業財經'
]
all_articles = open('all_articles.txt', 'w',encoding='utf-8')
#抓報社名稱
for newsPaperType in range(2601,2604):
    #抓該報紙的第一頁
    res = requests.get(base+date+str(newsPaperType))
    soup = BeautifulSoup(res.text, "lxml")
    #news
    news = soup.find('div',  {'class': 'listRight'})
    news = news.find_all('a')
    #不斷抓取下一頁資訊 直到沒有下一頁
    while True:
        for i in news:
            title = i.text.strip()
            if title in banlist:
                continue
            newsURL = i.get('href')
            #print(title)
            #print(newsURL)
            #all_articles.write(title + "," + base + newsURL + ",")
            #抓新聞內容
            newsContent = ""
            newsRes = requests.get(base+newsURL)
            newsSoup = BeautifulSoup(newsRes.text, "lxml")
            text = newsSoup.find('article', {'class': 'arttext marbotm clear-fix'})
            text = text.find_all('p')
            for i in text:
                #print(i.text)
                newsContent = newsContent + i.text
            if keyword in newsContent:
                print(newsContent)
                all_articles.write(title + ",")
                all_articles.write(newsContent)
                all_articles.write("\n")
        print('-------------------------------')
        #抓下一頁的網址
        nextPage = soup.find('div',  {'class': 'pagination clear-fix'})
        nextPage = nextPage.find_all('a')
        nextPageURL = ""
        for i in nextPage:
            if i.text == '下一頁':
                nextPageURL = i.get('href')
                #print(nextPageURL)
                break
        #print(base+nextPageURL)
        if nextPageURL == "":
            break
        time.sleep(0.1)
        res = requests.get(base+nextPageURL)
        soup = BeautifulSoup(res.text, "lxml")
        #news
        news = soup.find('div',  {'class': 'listRight'})
        news = news.find_all('a')
    
#print('-------------------------------')
#print('屌虐')
all_articles.close()
```
