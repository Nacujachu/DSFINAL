```python
import os

dir_name = 'CTdata_all'
file_names = os.listdir(dir_name)


keywords = ['柯文哲','一例一休','吳音寧','侯友宜','張安樂','蔡英文','墾丁']

for keyword in keywords:
    if not os.path.isdir(keyword+'_中時'):
        os.mkdir(keyword+'_中時')

for file in file_names:
    print(file)
    lines = open(dir_name + '/'+file,'r',encoding='utf-8').readlines()
    for keyword in keywords:
        write_file_name = keyword+'_中時'+ '/' + keyword + '-' + file
        out_f = open(write_file_name , 'w' , encoding='utf-8')
        
        for line in lines:
        
            if keyword in line:
                
                out_f.write(line)
        out_f.close()
            
    
    
    
    
```
